```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_table('titanic.txt')
```

```python
X.head()
```

```python
def classToInt(str):
    return int(str[0])
```

```python
X['PClass'] = X['PClass'].apply(classToInt)
```

```python
X.head()
```

```python
X.tail()
```

```python
groups = X.groupby('PClass')
print(len(groups))
observed = np.zeros((2,len(groups)), dtype=int)
expected = np.empty(observed.shape, dtype=int)
print(observed)
```

```python
print(observed)
for i in range(1,4):
    g = groups.get_group(i)
    for row in g:
        observed[0][i] += 1
```
